Vamos começar usando SciKitLearn para criar dados específicos para o modelo K-means.

Fonte deste notebook: https://jakevdp.github.io/PythonDataScienceHandbook/05.11-k-means.html

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

Por favor, usem os dados criados para treinar um modelo de clusterização k-means do scikit learn e depois tentem plotar os dados usando a a biblioteca matplotlib

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Finalmente, vamos verificar como ficou o resultado.

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
